### simple queries to explore the data in the tables

### top ten rated films, ordered in descending order by number of votes

```python
%%bigquery 
select * from imdb_staging.TitleRatings
where averageRating > 9.9
order by numVotes desc
limit 10
```

### crew members (identified by nconst) in the greatest number of roles

```python
%%bigquery
select nconst,
    count(*) AS `num`
    from imdb_staging.TitlePrincipals
    group by nconst
    order by num desc
    limit 10
```

### earliest people in the industry

```python
%%bigquery
select * from imdb_staging.NameBasics
order by string_field_2 asc
limit 10
```

### 10  oldest Adult Films

```python
%%bigquery 
select*from imdb_staging.TitleBasics
where isAdult = 1 and startYear > 0
order by startYear 
limit 10
```

### directors that have directed the greatest amount of works

```python
%%bigquery
select string_field_1, 
    count(*) AS `num`
    from imdb_staging.TitleCrew
    group by string_field_1
    order by num desc
    limit 10
```
