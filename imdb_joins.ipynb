## Highest rated movies (including genre description) with at least 100 votes

```python
%%bigquery
SELECT primaryTitle, averageRating
FROM imdb_modeled.Titles t
JOIN imdb_modeled.TitleRatings tr
ON t.tconst = tr.tconst
WHERE tr.numVotes > 100 AND t.titleType = "movie"
ORDER BY tr.averageRating DESC
LIMIT 10
```

## Worst rated TV Series (including genre description) with at least 100 votes

```python
%%bigquery
SELECT primaryTitle, averageRating
FROM imdb_modeled.Titles t
JOIN imdb_modeled.TitleRatings tr
ON t.tconst = tr.tconst
WHERE tr.numVotes > 100 AND t.titleType = "tvSeries"
ORDER BY tr.averageRating ASC
LIMIT 10
```

## Directors and their different roles that are my age

```python
%%bigquery
SELECT p.primaryName, p.primaryProfession
FROM imdb_modeled.People p
LEFT JOIN imdb_modeled.Directs d
ON p.primaryName = d.directors
WHERE p.birthYear = 1998
LIMIT 10
```

### Cast members of the oldest Titles

```python
%%bigquery
SELECT DISTINCT startYear, t.tconst, p.nconst, p.primaryName
FROM imdb_modeled.Titles AS t
FULL JOIN imdb_staging.TitlePrincipals AS tp 
ON t.tconst = tp.tconst
JOIN imdb_modeled.People p
ON tp.nConst = p.nConst
WHERE startYear > 0
ORDER BY startYear ASC
LIMIT 10
```

### Titles sorted by best rated with greater than 1,000,000 votes

```python
%%bigquery 
SELECT DISTINCT primaryTitle, averageRating, numVotes
FROM imdb_modeled.TitleRatings tr
FULL JOIN imdb_modeled.Titles t
ON t.tconst = tr.tconst
WHERE numVotes > 1000000
ORDER BY averageRating DESC
LIMIT 10
```

### Actors from the Office (US/UK)

```python
%%bigquery 
SELECT DISTINCT t.primaryTitle, p.primaryName, p.knownForTitles, t.startYear, t.runtimeMinutes 
FROM imdb_modeled.Titles t 
FULL JOIN imdb_staging.TitlePrincipals tp
ON tp.tConst = t.tConst
FULL JOIN imdb_modeled.People p
ON p.nConst = tp.nConst
WHERE primaryTitle = "The Office" AND t.startYear > 2005
ORDER BY t.runtimeMinutes
LIMIT 10
```
