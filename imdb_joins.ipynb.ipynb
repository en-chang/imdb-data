## Highest rated movies (including genre description) with at least 100 votes

```python
%%bigquery
SELECT primaryTitle, genres
FROM imdb_staging.TitleBasics tb
JOIN imdb_staging.TitleRatings tr
ON tb.tconst = tr.tconst
WHERE tr.numVotes > 100 AND tb.titleType = "movie"
ORDER BY tr.averageRating DESC
LIMIT 10
```

## Worst rated TV Series (including genre description) with at least 100 votes

```python
%%bigquery
SELECT primaryTitle, genres
FROM imdb_staging.TitleBasics tb
JOIN imdb_staging.TitleRatings tr
ON tb.tconst = tr.tconst
WHERE tr.numVotes > 100 AND tb.titleType = "tvSeries"
ORDER BY tr.averageRating ASC
LIMIT 10
```

## Directors and their different roles that are my age

```python
%%bigquery
SELECT nb.string_field_1, nb.string_field_4
FROM imdb_staging.NameBasics nb
LEFT JOIN imdb_staging.TitleCrew tc
ON nb.string_field_0 = tc.string_field_1
WHERE nb.string_field_2 = "1998"
LIMIT 10
```

### Cast members of the oldest Titles

```python
%%bigquery
select distinct startYear, tb.tconst, nconst, string_field_1
from imdb_staging.TitleBasics as tb full join imdb_staging.TitlePrincipals as tp on tb.tconst=tp.tconst
join imdb_staging.NameBasics nb on tp.nconst = nb.string_field_0
where startYear > 0
order by startYear asc
limit 10
```

### Titles sorted by best rated with greater than 1,000,000 votes

```python
%%bigquery 
select distinct averageRating, genres, primaryTitle, numVotes
from imdb_staging.TitleRatings tr full join imdb_staging.TitleBasics tb on tb.tconst = tr.tconst
where numVotes > 1000000
order by averageRating desc

limit 10
```

### Actors from the Office (US/UK)

```python
%%bigquery 
select distinct tb.primaryTitle, nb.string_field_1 as Name, tb.genres, tp.category, tb.startYear, tb.runtimeMinutes from imdb_staging.TitleBasics tb 
full join imdb_staging.TitlePrincipals tp on tp.tconst = tb.tconst
full join imdb_staging.NameBasics nb on nb.string_field_0 = tp.nconst
where primaryTitle = "The Office" and tb.startYear > 2005 and genres = "Documentary" 
order by tb.runtimeMinutes
limit 13
```
